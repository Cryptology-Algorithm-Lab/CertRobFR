```python
# Load Common Libraries
import torch
from torch import nn
import torch.nn.functional as F
import torchvision
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader
```

```python
# Setting Up the device
device = "cuda:0"
torch.backends.cudnn.benchmark = True
```

### 1. Experiments on Section 3.2

```python
# MNIST dataset

train_transforms = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
])

test_transforms = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
])

train_dataset_mnist = torchvision.datasets.MNIST(root = '', train = True, transform = train_transforms, download = True)
test_dataset_mnist = torchvision.datasets.MNIST(root = '', train = False, transform = test_transforms, download = True)


# CIFAR10 dataset
train_transforms = torchvision.transforms.Compose([
    torchvision.transforms.RandomCrop(32, padding = 4),
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.ToTensor(),
])

test_transforms = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
])

train_dataset_cifar = torchvision.datasets.CIFAR10(root = '.', train = True, transform = train_transforms, download = True)
test_dataset_cifar = torchvision.datasets.CIFAR10(root = '.', train = False, transform = test_transforms, download = True)
```

```python
# Backbones
from backbones.toyexamples import SLLNet_MNIST, SLLNet_CIFAR, Net_MNIST, Net_CIFAR

# A Simple Training Code
from toyexamples import train, benchmark
```

### Results on MNIST

```python
# Base Config
from easydict import EasyDict as edict
cfg = edict()
cfg.batch_size = 512
cfg.lr = 0.001
cfg.wd = 0
cfg.device = device 
cfg.epoch = 100
cfg.verbose = 20
cfg.train_dataset = train_dataset_mnist
cfg.test_dataset = test_dataset_mnist
cfg.xi = 0
cfg.lambd = 10
```

```python
# This may take a while... Be patient.
net, net_header, acc_net, avg_fn_net = train(cfg, Net_MNIST)
sllnet, sll_header, acc_sll, avg_fn_sll = train(cfg, SLLNet_MNIST)

# Enable xi
cfg.xi = 1
sllnet1, sll_header1, acc_sll1, avg_fn_sll1 = train(cfg, SLLNet_MNIST)
cfg.xi = 3
sllnet3, sll_header3, acc_sll3, avg_fn_sll3 = train(cfg, SLLNet_MNIST)
cfg.xi = 5
sllnet5, sll_header5, acc_sll5, avg_fn_sll5 = train(cfg, SLLNet_MNIST)
```

```python
# Plot the result
# Average Feature Norm
plt.figure(figsize = (6,6))
plt.title("MNIST", fontsize = 20, weight ="bold")
plt.plot(avg_fn_sll, "r", label = r"$\xi=0$", linewidth = 2)
plt.plot(avg_fn_sll1, "g", label = r"$\xi=1$", linewidth = 2)
plt.plot(avg_fn_sll3, "b", label = r"$\xi=3$", linewidth = 2)
plt.plot(avg_fn_sll5, "y", label = r"$\xi=5$", linewidth = 2)
plt.grid("on", linewidth = 1.5)
plt.xticks(np.linspace(0, 100, 5), fontsize = 20)
plt.yticks(np.linspace(0, 5, 6), fontsize = 20)
plt.ylabel("Avg. FN", fontsize = 30)
plt.legend(fontsize = 20)
plt.show()

# Accuracy
plt.figure(figsize = (6,6))
plt.title("MNIST", fontsize = 20, weight ="bold")
plt.plot(acc_sll, "r", label = r"$\xi=0$", linewidth = 2)
plt.plot(acc_sll1, "g", label = r"$\xi=1$", linewidth = 2)
plt.plot(acc_sll3, "b", label = r"$\xi=3$", linewidth = 2)
plt.plot(acc_sll5, "y", label = r"$\xi=5$", linewidth = 2)
plt.plot(acc_net, "k--", label = "Non.", linewidth = 2)
plt.grid("on", linewidth = 1.5)
plt.xticks(np.linspace(0, 100, 5), fontsize = 20)
plt.yticks(np.linspace(0.9, 1, 6), fontsize = 20)
plt.ylabel("Test ACC", fontsize = 30)
plt.ylim(0.9, 1.01)
plt.legend(fontsize = 20)
plt.show()
```

### 2. Visualization of Fig 1.(b)
- This reproduces a Fig.1 in Appendix C.1, Supplementary material.

```python
# Processing data...
test_loader = DataLoader(test_dataset_cifar, batch_size = 128, shuffle = False, num_workers = 4)
data, labels, feats = [], [], []

# Extract Features
with torch.no_grad():
    for (x, y) in test_loader:
        data.append(x); labels.append(y);
        feat = sllnet(x.to(device)).cpu()
        feats.append(feat)

data = torch.cat(data, dim= 0).numpy().reshape(-1, 3072)
labels = torch.cat(labels, dim = 0).numpy()
feats = torch.cat(feats, dim=0).numpy()  
```

```python
# We will use PCA for preserving the scale of the norm.
from sklearn.decomposition import PCA

# Do PCA on images
pca = PCA(n_components = 2)
pca = pca.fit(data)
data_pca = pca.fit_transform(data)

# Do PCA on feats
pca = PCA(n_components = 2)
pca = pca.fit(feats)
feat_pca = pca.fit_transform(feats)
```

```python
# Distribution of Images
plt.figure(figsize = (4,4))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c = labels, alpha = .4)
plt.title("Images", fontsize = 20, weight ="bold")
plt.xticks(np.linspace(-25, 30, 3), fontsize = 17.5, weight ="bold")
plt.yticks(np.linspace(-15, 20, 3), fontsize = 17.5, weight ="bold")
plt.show()

# Distribution of Features
plt.figure(figsize = (4,4))
plt.scatter(feat_pca[:, 0], feat_pca[:, 1], c = labels, alpha = .4)
plt.title("Features", fontsize = 20, weight ="bold")
plt.xticks(np.linspace(-0.1, 0.1, 3), fontsize = 17.5, weight ="bold")
plt.yticks(np.linspace(-0.1, 0.1, 3), fontsize = 17.5, weight ="bold")
plt.show()
```
