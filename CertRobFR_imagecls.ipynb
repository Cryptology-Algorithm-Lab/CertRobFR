### Read Logs

Caveat. This only works for the enclosed log in `logs`

```python
f = open("./logs/training_lln.log")
data = f.readlines()
f.close()

mem_lfw = []
mem_acc = []
mem_fn = []

for line in data:
    if "[lfw]" in line:        
        if "Norm" in line:
            mem_fn.append(line)
        elif "Accuracy-F" in line:
            mem_acc.append(line)
            
        mem_lfw.append(line)
        
mem_fndat = []
mem_accdat = []

for i in range(101):
    dat_fn = mem_fn[i][-10:-1].replace(" ", "")
    dat_acc = mem_acc[i][-17:-10].replace(" ", "")
    mem_fndat.append(float(dat_fn))
    mem_accdat.append(float(dat_acc))
    
mem_loss = []

for line in data:
    if "Loss" in line:
        mem_loss.append(line)    
        
mem_lossdat = []
for dat in mem_loss:
    idx = dat.index("Loss ")
    datt = dat[idx+4:idx+11].replace(" ", "")
    mem_lossdat.append(float(datt))
            
steps = [i * 2000 for i in range(1, 102)]        
step_loss = [20 * i for i in range(1, 10116)]        
```

### Visualization

```python
plt.figure(figsize=(6, 6))
plt.ylabel("Train Loss", fontsize = 30)
plt.plot(step_loss, mem_lossdat, 'r', linewidth = 2)
plt.xticks(np.linspace(0, 200000, 6), fontsize = 20)
plt.yticks(np.linspace(10, 22, 5), fontsize = 20)
plt.ticklabel_format(axis='x', style='sci', scilimits=(1,4))
plt.grid("on")
plt.show()
```

```python

plt.figure(figsize=(6, 6))
plt.ylabel("Avg. FN", fontsize = 30)
plt.plot(steps, mem_fndat, 'b', linewidth = 2)
plt.plot(steps, [13.07] * len(steps), "k--", linewidth = 2)
plt.xticks(np.linspace(0, 200000, 6), fontsize = 20)
plt.yticks(np.linspace(5, 20, 6), fontsize = 20)
plt.ticklabel_format(axis='x', style='sci', scilimits=(1,4))

plt.grid("on")
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(6, 6))
plt.ylabel("Test ACC", fontsize = 30)
plt.plot(steps, mem_accdat, 'g', linewidth = 2)
plt.xticks(np.linspace(0, 200000, 6), fontsize = 20)
plt.yticks(np.linspace(0.79, 0.87, 5), fontsize = 20)
plt.ticklabel_format(axis='x', style='sci', scilimits=(1,4))
plt.grid("on")
plt.show()
```
